```python
import pandas as pd
from textaugment import Wordnet

# Initialize Wordnet augmenter
t = Wordnet()

def augment_text(text, num_augmentations=3):
    augmented_texts = [t.augment(text) for _ in range(num_augmentations)]
    return augmented_texts

# Apply augmentation
df['Augmented_Prompts'] = df['Prompt'].apply(augment_text)

# Exploding the DataFrame to have separate rows for each augmented text
df = df.explode('Augmented_Prompts')

# Display the first few rows of the augmented DataFrame
print(len(df))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Embedding, Flatten
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical


# Load GloVe embeddings
def load_glove_embeddings(glove_file):
    embeddings_index = {}
    with open(glove_file, 'r', encoding='utf8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_index[word] = coefs
    return embeddings_index

glove_embeddings = load_glove_embeddings('glove.6B.100d.txt')  # Update path

# Tokenize and pad sequences
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['Prompt'])
sequences = tokenizer.texts_to_sequences(df['Prompt'])
word_index = tokenizer.word_index
# max_length = max(len(seq) for seq in sequences)
max_length = 150
data_padded = pad_sequences(sequences, maxlen=max_length)

# Create embedding matrix
embedding_dim = 100
embedding_matrix = np.zeros((len(word_index) + 1, embedding_dim))
for word, i in word_index.items():
    embedding_vector = glove_embeddings.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector

# Labels
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(df['Class'])
labels = to_categorical(labels)

# Splitting the dataset
# X_train, X_test, y_train, y_test = train_test_split(data_padded, labels, test_size=0.3, random_state=42)
X = data_padded
y = labels

# DNN Model with Embedding Layer
model = Sequential()
model.add(Embedding(len(word_index) + 1, embedding_dim, input_length=max_length, weights=[embedding_matrix], trainable=False))
model.add(Flatten())  # Flatten the output to feed into Dense layers
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(y.shape[1], activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Model training
model.fit(X, y, epochs=50, batch_size=5, validation_data=(X, y))

# Function to predict the class of a new prompt
def predict_class(new_prompt):
    new_seq = tokenizer.texts_to_sequences([new_prompt])
    new_padded = pad_sequences(new_seq, maxlen=max_length)
    prediction = model.predict(new_padded)
    return label_encoder.inverse_transform([np.argmax(prediction)])

# Example usage
new_prompt = "Why are noodles so important? Explain me with just image."
predicted_class = predict_class(new_prompt)
print("Predicted Class:", predicted_class[0])
```

```python
import pickle
# Save the model
model.save('my_model.h5')

json_string = model.to_json()
with open('model_architecture.json', 'w') as json_file:
    json_file.write(json_string)

# Load the tokenizer
with open('tokenizer.pkl', 'wb') as handle:
    pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)

# Load the label encoder
with open('label_encoder.pkl', 'wb') as handle:
    pickle.dump(label_encoder, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
